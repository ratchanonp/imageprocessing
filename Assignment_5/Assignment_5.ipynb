```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Read image
fruite_path = "img/fruit.jpg"
fruite = cv.imread(fruite_path)
```

```python
# Write A function to create complementary color image
def complementary_color(img):
    # Split the image into three channels
    b, g, r = cv.split(img)
    # Create complementary color image
    b = 255 - b
    g = 255 - g
    r = 255 - r
    # Merge three channels
    img = cv.merge([b, g, r])
    return img
```

```python
# Remove the output folder if it exists
if os.path.exists("output/"):
    os.system("rm -rf output/")

os.mkdir("output/")

if not os.path.exists("output/5_1"):
    os.mkdir("output/5_1")

# Write the image to output folder
complementary_frutie = complementary_color(fruite)
cv.imwrite("output/5_1/complementary_fruite.jpg", complementary_frutie)

# Show original image and complementary color image
plt.figure(figsize=(10, 10))
plt.subplot(121)
plt.title("Original Image")
plt.imshow(cv.cvtColor(fruite, cv.COLOR_BGR2RGB))
plt.axis("off")
plt.subplot(122)
plt.title("Complementary Color Image")
plt.imshow(cv.cvtColor(complementary_frutie, cv.COLOR_BGR2RGB))
plt.axis("off")
plt.savefig("output/5_1/complementary_fruite_plot.png")
plt.show()
```
